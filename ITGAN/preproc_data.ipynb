```python
import pandas as pd
import numpy as np

data = np.load('util/data/malware.npz')
```

```python
data.items()
```

```python
# Get unique values of each column and put them in a dictionary
i2s= {}
size = {}
for column in data.columns:
    if data[column].dtype in ('object','int64'):
        i2s[column] = list(data[column].unique())
        size[column] = list(data[column].unique()).__len__()
    else:
        i2s[column] = pd.NA
        size[column] = pd.NA
```

```python
size
```

```python
# get min and max values of each column and put them in a dictionary
min = {}
max = {}
for column in data.columns:
    if data[column].dtype not in ('object','int64'):
        min[column] = data[column].min()
        max[column] = data[column].max()
    else:
        min[column] = pd.NA
        max[column] = pd.NA
```

```python
max
```

```python
# type of each column
type = {}
for column in data.columns:
    if data[column].dtype in ('object','int64'):
        type[column] = "categorical"
    else:
        type[column] = "continuous"
```

```python
max.values()
```

```python
# create a pandas dataframe with the information
info = pd.DataFrame({'max': max.values(), 'min': min.values(), 'name': max.keys(), 'type': type.values(), 'i2s': i2s.values(), 'size': size.values()})
```

```python
info = info.to_dict("records")
```
